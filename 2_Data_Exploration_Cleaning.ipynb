### Remove currency symbol from currency data

```python
df[df['price'].str.startswith('$')]
```

```python
currency_columns = [ 'price',
# 'weekly_price',
# 'monthly_price',
 'security_deposit',
 'cleaning_fee',
 'extra_people']
```

```python
for col in currency_columns:
    df[f'{col}'] = df[f'{col}'].replace('\$','', regex=True)
    df[f'{col}'] = df[f'{col}'].replace(',','', regex=True)
    df[f'{col}'] = df[f'{col}'].astype(float)
    df[f'{col}'] = df[f'{col}'].fillna(0)
    
```
